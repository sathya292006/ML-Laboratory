```python
# STEP 1: Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA


# STEP 2: Create dataset (10 students, 2 attributes)
data = {
    'Internal':   [70, 60, 85, 75, 90, 65, 88, 72, 95, 78],
    'FinalExam': [75, 68, 92, 80, 94, 70, 90, 76, 98, 82]
}

df = pd.DataFrame(data)

print("Original Dataset:")
print(df)


# STEP 3: Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)


# STEP 4: Apply PCA
pca = PCA()
pca.fit(scaled_data)


# STEP 5: Print Explained Variance Ratio
print("\nExplained Variance Ratio:")
print(pca.explained_variance_ratio_)


# STEP 6: Scree Plot
plt.figure()
plt.plot(
    range(1, len(pca.explained_variance_ratio_) + 1),
    pca.explained_variance_ratio_,
    marker='o'
)
plt.xlabel("Principal Component Number")
plt.ylabel("Explained Variance Ratio")
plt.title("Scree Plot")
plt.show()


# STEP 7: Reduce data to 1 Principal Component
pca_1 = PCA(n_components=1)
reduced_data = pca_1.fit_transform(scaled_data)

print("\nReduced Data (PC1):")
print(reduced_data)
```
