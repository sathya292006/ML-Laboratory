```python
# ================================
# PCA LAB – GOOGLE COLAB CODE
# ================================

# Step 1: Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Step 2: Create dataset (10 examples)
data = {
    'Internal':   [70, 60, 85, 75, 90, 65, 88, 72, 95, 78],
    'FinalExam': [75, 68, 92, 80, 94, 70, 90, 76, 98, 82]
}

df = pd.DataFrame(data)

print("Original Dataset:")
print(df)

# Step 3: Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# Step 4: Apply PCA
pca = PCA()
pca.fit(scaled_data)

print("\nExplained Variance Ratio:")
print(pca.explained_variance_ratio_)

# Step 5: Reduce data to 1 Principal Component
pca_1 = PCA(n_components=1)
reduced_data = pca_1.fit_transform(scaled_data)

print("\nReduced Data (PC1):")
print(reduced_data)

# Step 6: Scatter Plot
plt.figure(figsize=(8,4))
plt.scatter(reduced_data, [0]*len(reduced_data))
plt.xlabel("Principal Component 1")
plt.title("PCA Scatter Plot (2 Features → 1 Feature)")
plt.yticks([])
plt.grid(True)
plt.show()
```
